```python
import pandas as pd

#Q1: Create two small DataFrames (Employees and Departments) and merge them using pd.merge() on a common key.

Employees = pd.DataFrame({
    "EMP_ID":[1,2,3,4],
    "Name":["Rohit","Vaishnavi","Ganesh","Vaibhav"],
    "Dept_ID":[101,102,103,101]
})

Departments = pd.DataFrame({
    "Dept_ID":[101,102,103],
    "Department":["HR","Sales","Marketing"]
})

merged = pd.merge(Employees,Departments,on="Dept_ID")

merged
```

```python
#Q2: Perform an inner join, left join, right join, and outer join between two DataFrames, and explain the difference in results.


# Employees DataFrame
employees = pd.DataFrame({
    'Emp_ID': [101, 102, 103, 104],
    'Name': ['Rohit', 'Amit', 'Priya', 'Sneha'],
    'Dept_ID': [1, 2, 3, 4]
})

# Departments DataFrame
departments = pd.DataFrame({
    'Dept_ID': [1, 2, 3, 5],
    'Department': ['HR', 'IT', 'Finance', 'Marketing']
})

employees
```

```python
departments
```

```python
# Inner Join

inner_join = pd.merge(employees,departments,on="Dept_ID",how="inner")

inner_join
```

```python
# Left Join

left_join = pd.merge(employees,departments,on="Dept_ID",how="left")

left_join
```

```python
right_join = pd.merge(employees,departments,on="Dept_ID",how="right")

right_join
```

```python
outer_join = pd.merge(employees,departments,on="Dept_ID",how = "outer")

outer_join
```

```python
#Q3: Concatenate two DataFrames with the same columns vertically using pd.concat().

# First DataFrame
df1 = pd.DataFrame({
    'Emp_ID': [101, 102],
    'Name': ['Rohit', 'Amit'],
    'Department': ['HR', 'IT']
})

# Second DataFrame
df2 = pd.DataFrame({
    'Emp_ID': [103, 104],
    'Name': ['Priya', 'Sneha'],
    'Department': ['Finance', 'Marketing']
})

result = pd.concat([df1,df2])

result
```

```python
#Q4: Concatenate two DataFrames side by side (horizontally) using pd.concat(axis=1).

# First DataFrame
df1 = pd.DataFrame({
    'Emp_ID': [101, 102],
    'Name': ['Rohit', 'Amit'],
    'Department': ['HR', 'IT']
})

# Second DataFrame
df2 = pd.DataFrame({
    'Emp_ID': [103, 104],
    'Name': ['Priya', 'Sneha'],
    'Department': ['Finance', 'Marketing']
})

result =  pd.concat([df1,df2],axis = 1)

result
```

```python
#Q5: Use join() to combine two DataFrames based on their index.

# First DataFrame
employees = pd.DataFrame({
    'Name': ['Rohit', 'Amit', 'Priya'],
    'Department': ['HR', 'IT', 'Finance']
}, index=[101, 102, 103])

# Second DataFrame
salary = pd.DataFrame({
    'Salary': [40000, 50000, 45000]
}, index=[101, 102, 103])


result = employees.join(salary)

result
```

```python
#Q6: Merge two DataFrames on multiple key columns (e.g., Region and Year).

# Sales DataFrame
sales = pd.DataFrame({
    'Region': ['North', 'South', 'East', 'West'],
    'Year': [2023, 2023, 2024, 2024],
    'Sales': [50000, 60000, 55000, 70000]
})

# Profit DataFrame
profit = pd.DataFrame({
    'Region': ['North', 'South', 'East', 'Central'],
    'Year': [2023, 2023, 2024, 2024],
    'Profit': [10000, 12000, 11000, 9000]
})

result = pd.merge(sales,profit,on=["Region","Year"],how='outer')

result
```

```python
#Q7: Handle overlapping column names during a merge using the suffixes parameter.

employees = pd.DataFrame({
    'Emp_ID': [101, 102, 103],
    'Name': ['Rohit', 'Amit', 'Priya'],
    'Salary': [40000, 50000, 45000]
})

# Second DataFrame
bonus = pd.DataFrame({
    'Emp_ID': [101, 102, 103],
    'Salary': [5000, 6000, 5500],   # Overlapping column name
    'Bonus': [2000, 2500, 2200]
})

result = pd.merge(employees,bonus,on="Emp_ID",suffixes=["_employee","_bonus"])

result
```

```python
#Q9: Identify rows that exist in one DataFrame but not the other using an outer join with an indicator column.

# First DataFrame
employees = pd.DataFrame({
    'Emp_ID': [101, 102, 103, 104],
    'Name': ['Rohit', 'Amit', 'Priya', 'Sneha']
})

# Second DataFrame
new_employees = pd.DataFrame({
    'Emp_ID': [103, 104, 105, 106],
    'Name': ['Priya', 'Sneha', 'Rahul', 'Neha']
})

result = pd.merge(employees,new_employees,on=["Emp_ID","Name"],how="outer",indicator="True")

result

```

```python
#Q10: Combine three or more DataFrames from different sources into a single master DataFrame.

# DataFrame 1: Employee Details
employees = pd.DataFrame({
    'Emp_ID': [101, 102, 103],
    'Name': ['Rohit', 'Amit', 'Priya']
})

# DataFrame 2: Department Details
departments = pd.DataFrame({
    'Emp_ID': [101, 102, 103],
    'Department': ['HR', 'IT', 'Finance']
})

# DataFrame 3: Salary Details
salary = pd.DataFrame({
    'Emp_ID': [101, 102, 103],
    'Salary': [40000, 50000, 45000]
})

result = pd.merge(employees,departments,on="Emp_ID")
result = pd.merge(result,salary,on="Emp_ID")

result
```

```python
#Q11: Merge a small lookup table (e.g., product ID to product name) into a larger sales DataFrame.

# Large Sales DataFrame
sales = pd.DataFrame({
    'Sale_ID': [1001, 1002, 1003, 1004, 1005],
    'Product_ID': [101, 102, 103, 101, 104],
    'Quantity': [2, 1, 5, 3, 4],
    'Amount': [1000, 1500, 2500, 1500, 4000]
})

# Small Lookup Table
products = pd.DataFrame({
    'Product_ID': [101, 102, 103, 104],
    'Product_Name': ['Laptop', 'Mouse', 'Keyboard', 'Monitor']
})

# Merge the lookup table into the sales DataFrame
master_sales = pd.merge(sales, products, on='Product_ID', how='left')

master_sales
```

```python
#Q12: Use merge_ordered() or merge_asof() to combine time-series data from two sources.

# Sales DataFrame
sales = pd.DataFrame({
    'Date': ['2024-01-01', '2024-01-03', '2024-01-05'],
    'Sales': [1000, 1500, 1200]
})

# Expenses DataFrame
expenses = pd.DataFrame({
    'Date': ['2024-01-02', '2024-01-03', '2024-01-06'],
    'Expense': [500, 700, 400]
})

sales["Date"] = pd.to_datetime(sales['Date'])
expenses["Date"] = pd.to_datetime(expenses['Date'])

sales.dtypes

expenses.dtypes

result = pd.merge_ordered(sales,expenses,on="Date")

result
```

```python
#Q13: After merging, handle any resulting missing values appropriately.

# Employee DataFrame
employees = pd.DataFrame({
    'Emp_ID': [101, 102, 103, 104],
    'Name': ['Rohit', 'Amit', 'Priya', 'Sneha'],
    'Dept_ID': [1, 2, 3, 4]
})

# Department DataFrame
departments = pd.DataFrame({
    'Dept_ID': [1, 2, 3],
    'Department': ['HR', 'IT', 'Finance']
})

result = pd.merge(employees,departments,on="Dept_ID",how="left")

result

```

```python
#Q15: Combine everything: merge Sales, Products, and Customers tables into one clean report and export it to a new CSV.


# Sales DataFrame
sales = pd.DataFrame({
    'Sale_ID': [1001, 1002, 1003, 1004],
    'Customer_ID': [1, 2, 3, 1],
    'Product_ID': [101, 102, 103, 104],
    'Quantity': [2, 1, 3, 2]
})

# Products DataFrame
products = pd.DataFrame({
    'Product_ID': [101, 102, 103, 104],
    'Product_Name': ['Laptop', 'Mouse', 'Keyboard', 'Monitor'],
    'Price': [50000, 800, 1500, 12000]
})

# Customers DataFrame
customers = pd.DataFrame({
    'Customer_ID': [1, 2, 3],
    'Customer_Name': ['Rohit', 'Amit', 'Priya'],
    'City': ['Pune', 'Mumbai', 'Nagpur']
})


report = pd.merge(sales,products,on="Product_ID",how="left")

report = pd.merge(report,customers,on="Customer_ID",how="left")

report
```

```python
report["Total_Amount"] = report["Quantity"] * report["Price"]
```

```python
report = report[['Sale_ID', 'Customer_Name', 'City',
                 'Product_Name', 'Quantity', 'Price', 'Total_Amount']]
```
