```python
#Q1: Create a NumPy array of 100 random numbers and calculate all key statistics (mean, median, std, min, max) in one script.

import numpy as np

import pandas as pd

array = np.random.randint(0,500,100)



mean = np.mean(array)
median = np.median(array)
std = np.std(array)
min = np.min(array)
max = np.max(array)

print("\n--- Statistics ---")
print("Mean   :", mean)
print("Median :", median)
print("Standard Deviation :", std)
print("Minimum:", min)
print("Maximum:", max)
```

```python
#Q3: Clean the Titanic dataset: handle missing Age and Embarked values, drop irrelevant columns.

titanic
```

```python
titanic["age"] = titanic["age"].fillna(titanic["age"].median)

titanic["embark_town"] = titanic["embark_town"].fillna(titanic["embark_town"].mode()[0])
```

```python
titanic.drop(columns=["who","deck"],inplace=True)
```

```python
titanic["embarked"] = titanic["embarked"].fillna(titanic["embarked"].mode)
```

```python
titanic.isnull().sum()
```

```python
#Q4: Find the survival rate (%) by gender using groupby().

survival_rate = titanic.groupby("sex")["survived"].mean() *100

survival_rate
```

```python
#Q5: Find the survival rate (%) by passenger class (Pclass) using groupby().

survival_rate = titanic.groupby("pclass")["survived"].mean()*100

survival_rate
```
