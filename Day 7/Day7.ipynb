```python
# Q2: Read a file line by line and print each with line number.

file = open("sample.txt","r")

line_number = 1

for line in file:
    print("Line",line_number,":",line.strip())
    line_number += 1

file.close()
```

```python
#Q4: Handle FileNotFoundError, ValueError, TypeError in single try-except block.

try:
    file = open("sample.txt","r")

    num = int(input("Enter a number :"))

    result = num + "python"

    print(result)

    file.close()

except FileNotFoundError:
    print("Error ! file not found.")

except ValueError:
    print("Error ! please enter a valid integer.")

except TypeError:
    print("Error ! Operation between two diff data types is not possible.")
```

```python
#Q6: Write employee records (name, ID, salary) to file. Read and display formatted.

file = open("employee.txt","w")

file.write("Rohit,1001,950000,Pune\n")
file.write("Vaishu,1002,100000,Pune\n")
file.write("Ganesh,1003,900000,Jalna\n")

file.close()

file = open("employee.txt","r")

print("Empoyee Records")
print("______________________________________________")

print("name\t  emp_id\t salary\t         city")

print("______________________________________________")

for line in file:
    name,emp_id,salary,city = line.strip().split(",")
    print(name,"\t  ",emp_id,"\t",salary,"\t",city)


file.close()
```

```python
#Q7: Write try-except-finally block: open file, read, handle errors, close properly


try:
    file = open("sample.txt","r")

    print(file.read())

except FileNotFoundError:
    print("Error ! File not found.")

finally:
    if file :
        file.close()
        print("File closed successfully.")

```

```python
#Q8: Validate user input (integer, age > 0) using try-except. Store in file.

try:
    age = int(input("Enter Your Age :"))

    print("Age you enter :",age)
    
    if age <= 0:
        print("Age must be greter than 0")

    else:
        file = open("age.txt","w")
        file.write("Age :" )
        file.close()
        print("Age stored successfully.")

except :
    print("Enter a valid integer.")
```
