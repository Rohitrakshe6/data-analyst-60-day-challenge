```python
# Matrix Operations

import numpy as np

A = np.array([[[1,2,3],
             [5,6,7],
             [4,7,2]]])

B = np.array([[[5,2,9],
             [1,2,4],
             [3,8,2]]])

print("Matrix A is :")
print(A,"\n")

print("Matrix B is :")
print(B,"\n")

```

```python
print("Addition :")
print(A+B)
```

```python
print("Subtraction :")
print(A-B)
```

```python
print("Element wise Multiplication :")
print(A*B)
```

```python
print("Matrix wise Multiplication :")
print(A@B)
```

```python
print("Transpose :")
print(A.T)
```

```python
print("Max :")
print(A.max(),"\n")

print("Min :")
print(A.min())
```

```python
print("Sum of Elements :",A.sum())
```

```python
print("Average  :",A.mean())
```

```python
print("Row wise Maximum :")
print(A.max(axis = 1))
```

```python
print("Column wise Maximum :")
print(A.max(axis = 0))
```

```python
print("Reshape :")
print(A.reshape(1,9))
```

```python
print("Flatten :")
print(A.flatten())
```
