```python
#Q2: Demonstrate dict methods: keys(), values(), items(), get(), pop().

student = {

    "Name" : "Rohit Rakshe",
    "Age" : 22,
    "Grade" : "A",
    "Marks" : 89

}

print("Keys")
print(student.keys())

print("\nValues")
print(student.values())

print("\nItems")
print(student.items())

print("\nGet")
print(student.get("Name"))

print("\nPop")
print(student.pop("Age"))

print(student)
```

```python
#Q4: Create two sets and perform union, intersection, difference operations.

set1 = {10, 20, 30, 40, 50}
set2 = {30, 40, 50, 60, 70}

print("Union")
print(set1.union(set2))

print("\nIntersection")
print(set1.intersection(set2))

print("\nDifference")
print(set1.difference(set2))

print("\nDifference")
print(set2.difference(set1))
```

```python
#Q8: Check if a string is a palindrome (ignore spaces and case)

text = str(input("Enter a Text Here :"))

if text == text[::-1]:
    print("It is a Palindrome")

else:
    print("Not a Palindrome")
    
```

```python
#Q9: Count word frequency in a sentence using a dictionary.

sentence = input("Enter a sentence: ")

words = sentence.lower().split()

frequency = {}

for word in words:
    if word in frequency:
        frequency[word] +=1
    
    else:
        frequency[word] =1

for key ,value in frequency.items():
    print(key ,":",value)
```

```python
#Q14: Check if two strings are anagrams.

text1 =str(input("Enter a First text here :"))

text2 =str(input("Enter a Second text here :"))

text1 = text1.replace("","").lower()

text2 = text2.replace("","").lower()

if sorted(text1) == sorted(text2):
    print("this two strings are anagrams.")

else:
    print("this two strings are not anagrams.")
```
