```python
balance = 100000

def check_balance():
    print("Current Balance is:",balance)

def deposite(amount):
    global balance
    balance = amount + balance
    print("Amount Deposited successfully.")
    print("current Balance is :",balance)

def withdraw(amount):
    global balance
    if amount <= balance :
        balance = balance - amount
        print("Please Collect Your Cash.")
        print("Current Balance is :", balance)
    else:
        print("Insufficient Balance")


while True:
    print("\n------ATM Menue------" )
    print("1. for checking Balance")
    print("2. for Deposite Amount")
    print("3. for Withdraw Amount")
    print("4. for Exit")
    print("\n---------------------" )

    choice = int(input("Enter Your Choice"))

    if choice == 1:
        (check_balance())

    elif choice == 2:
        amount = int(input("Enter Amount want to Deposite :"))
        deposite(amount)

    elif choice ==3:
        amount = int(input("Enter amoubnt want to Withdraw :"))
        withdraw(amount)       

    elif choice == 4:
        print("Thank You ")
        break

    else:
        print("Enter valid choice")
```
