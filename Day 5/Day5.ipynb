```python
#Q4: Write a function to calculate sum and average of list elements.

def calculate(first):
    total = sum(first)
    average = total/len(first)
    return total,average

numbers = [10,20,30,40,50]

total,average = calculate(numbers)

print(total,average)
```

```python
#Q8: Find highest, lowest, and average marks from a list. Count students with marks > 80.

marks = [85, 72, 90, 65, 88, 78, 95]

print("Highest Marks :",max(marks))
print("Lowest Marks :",min(marks))
print("Average Marks :",sum(marks)/len(marks))

count = 0 

for i in marks:
    if i > 80:
        count +=1

print("Students Count Above 80 Marks is :",count)
```

```python
#Q11: Remove duplicates from a list while preserving order.

list = [1,2,3,3,4,2,5,6,7]

result =[]

for i in list :
    if i not in result :
        result.append(i)


print("Original List is :",list)

print("List after removing duplicates is :",result)
```

```python
#Q12: Flatten a nested list [[1,2], [3,4], [5,6]].

nested_list =[[1,2], [3,4], [5,6]]

flatten_list = []

for sub_list in nested_list:
    for item in sub_list :
        flatten_list.append(item)

print(flatten_list)
```

```python
#Q20: Move all zeros to the end of a list.

list = [1,2,0,4,3,0,6]

result = []

for i in list :
    if i != 0 :
        result.append(i)


for i in list :
    if i == 0 :
        result.append(i)


print(result)
```

```python
#Q25: Find intersection and union of two list
a = [1,2,3,4,5]

b = [1,4,7,8,9]

result = []

for i in a :
    for j in b:
        if i == j:
            result.append(j)

print(result)

union = list(set(a) | set(b))

print(union)
```
