```python
#Grade Calculator

name = str(input("Enter Students name :"))

print("Students name :",name)

english = int(input("Enter English Marks :"))

print("English Marks :",english)

math = int(input("Enter Mathmatics Marks :"))

print("Mathmatics Marks :",math)

science = int(input("Enter Science Marks :"))

print("Science Marks :",science)

marathi = int(input("Enter Marathi Marks :"))

print("Marathi Marks :",marathi)

hindi = int(input("Enter hindi Marks :"))

print("Hindi Marks :",hindi)

total = english + math + science + marathi + hindi

percentage = total / 5



print("\n========== REPORT CARD ==========")
print("Student Name :", name)
print("English      :", english)
print("Mathematics  :", math)
print("Science      :", science)
print("Marathi  :", marathi)
print("Hindi    :", hindi)
print("---------------------------------")
print("Total Marks  :", total, "/500")
print("Percentage   :", percentage, "%")
if percentage > 90 :
    print("Grade A+")

elif percentage >= 80 :
     print("Grade A")

elif percentage > 70 :
    print("Grade B")

elif percentage > 60 :
    print("Grade c")

elif percentage >= 40 :
    print("Grade D")

else:
    print("Grade F")
print("=================================")

```
