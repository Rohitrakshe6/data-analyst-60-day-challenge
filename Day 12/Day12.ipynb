```python
#Q1: Load the Titanic dataset (or any CSV with missing values) and count missing values in each column using isnull().sum()

import pandas as pd
import numpy as np

data = pd.read_csv(r"C:\Users\Dell\OneDrive\Desktop\Data Analyst 2.0\Python Projects\Day 12\employee_dirty_dataset_1000_rows.csv")
```

```python
data
```

```python
data.isnull().sum()
```

```python
#Q2: Drop rows that contain any missing values using dropna().

dropna_data = data.dropna()

dropna_data
```

```python
dropna_data.isnull().sum()
```

```python
#Q3: Fill missing numeric values with the column mean using fillna().

data.head(50)

data = data.fillna(data.mean(numeric_only=True))

data.tail(50)
```

```python
#Q4: Fill missing categorical values with the most frequent value (mode).

data = data.fillna(data.mode(numeric_only=True))

data
```

```python
#Q5: Detect and remove duplicate rows from a DataFrame using duplicated() and drop_duplicates().

data.duplicated()

data.duplicated().sum()

data = data.drop_duplicates()
```

```python
data.duplicated().sum()
```

```python
data.shape
```

```python
#Q6: Convert a column's data type (e.g., string to numeric) using astype() or pd.to_numeric()

data.dtypes
```

```python
data["Age"] = data["Age"].fillna(data["Age"].mean()).astype(int)

data.dtypes
```

```python
data["JoinDate"] = pd.to_datetime(data["JoinDate"],errors = "coerce")
```

```python
data.dtypes
```

```python
#Q8: Standardize inconsistent text data in a column (e.g., 'male', 'Male', 'MALE' -> 'Male').

#str.strip() → Removes extra spaces.
#str.capitalize() → Converts the first letter to uppercase and the rest to lowercase.

data["Gender"] = data["Gender"].str.strip().str.capitalize()

data
```

```python
#Q10: Replace specific values in a column using replace() (e.g., replace 'NaN' strings with actual NaN).

data["Department"] = data["Department"].replace(np.nan,"Unknown")

data

```

```python
#Q11: Create a new column that flags rows with missing values (1 if any missing, else 0)

data["Missing_flag"] = data.isnull().any(axis=1).astype(int)
```

```python
data
```

```python
#Q13: Bin a continuous numeric column into categories using pd.cut() (e.g., Age into child/adult/senior).

data["Age_Group"] = pd.cut(data["Age"],bins=[0,18,40,60],labels = ["Child","Adult","Senior"])

```

```python
print(data[["Age","Age_Group"]])
```

```python
data["JoinDate"] = data["JoinDate"].fillna("2023-01-01")
```

```python
#Q15: Combine all the above techniques to fully clean a raw dataset and export the cleaned version to a new CSV.

data.to_csv("Employee_Cleaned_dataset",index=False)

print("Cleaned Dataset Exported Successfully")
```
