```python
num1 = float(input("Enter Num 1 :"))

operation = input("Enter Operator (+,-,*,/) :")

num2 = float(input("Enter Num 2 :"))

if operation == "+":
    print("Result :",num1+num2)

elif operation == "-":
    print("Result :", num1-num2)

elif operation == "*":
    print("Result :", num1*num2)

elif operation == "/":
    if num2 != 0:
        print("Result :", num1/num2)

    else:
        print("Division by zero is not possible :")

else:
    print("Enter valid operator :")



```
