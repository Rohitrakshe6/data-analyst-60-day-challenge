```python
#Q8: Create a BMI calculator: BMI = weight / (height²). Classify as underweight/normal/overweight.

Weight = float(input("Enter your Weight in KG :"))

Height = float(input("Enter your Height in Meter :"))

BMI = Weight/(Height*Height)

print("BMI:",BMI)

if BMI < 18.5:
    print("Underweight")
elif BMI > 18.5 and BMI < 24.9 :
    print("Normal")
else: 
    print("Overweight")
```

```python
#Q10: Calculate simple interest: SI = (P × R × T) / 100.

"""
P = Principal Amount → The original amount of money borrowed or invested.
R = Rate of Interest → The annual interest rate (in percent, %).
T = Time → The duration for which the money is borrowed or invested (usually in years).
"""

P = float(input("Enter the Amount invested or borrowed :"))

R = float(input("Enter Rate of intrest in percentage :"))

T = float(input("Enter the duration of money invested or borrowed in years:"))

print("principle Amount:",P,"Rs")

print("Rate of Interest :",R,"%")

print("Time :",T,"Years")

Simple_Interest = (P * R * T) / 100

print("Simple Interest :",Simple_Interest,"Rs")

```

```python
#Q19: Find the compound interest using CI = P(1 + R/100)^T.

P = float(input("Enter Principal Amount (Original Money) :"))

R = float(input("Enter Rate of Interest (per year) in percentage :"))

T =float(input("Enter Time (in years) :"))

compound_interest = P*(1 + (R/100))**T

print("Principal Amount (Original Money) is :",P,"Rs")

print("Rate of Interest (per year) is :",R,"%")

print("Time (in years) is :",T,"Years")

print("Comound Interest is :",compound_interest)
```
