```python
# Task : Pattern Printing 
```

# 1. Right Triangle

```python
for i in range(1,8):
    print("*" * i)
```

# 2.Inverted Left Triangle

```python
for i in range(8,1,-1):
    print("*" * i)
```

# 3. Square Pattern

```python
for i in range(5):
    print("* " * 5)
```

# 4.Pyramid

```python
for i in range(1,6):
    print(" " * (5-i) + "* " * i )
```

# 5.Inverted Pyramid

```python
for i in range(6,0,-1):
    print( " " * (6-i) + "* " * i )
```

# 6. Number Triangle

```python
for i in range (1,6):
    for j in range (1,i+1):
      print(j,end = " ")
    print()
```

# 7.Inverted Triangle


```python
for i in range (6,0,-1):
    for j in range (1,i+1):
      print(j,end = " ")
    print()
```

# 8 Diamond Pattern

```python
for i in range(1,6):
    print(" " * (5-i) + "* " * i )

for i in range(5,0,-1):
    print(" "  * (5-i) + "* " * i )
```
