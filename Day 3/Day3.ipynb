```python
#Q1: Print numbers from 1 to 10 using for loop with range()

for i in range (1,11):
    print(i)
```

```python
#Q2: Print multiplication table of a number (1-10).

num = int(input("Enter Number : "))

for i in range (1,11):
    print(num*i)
```

```python
#Q3: Calculate sum of first N natural numbers using while loop.

n = int(input("Enter number : "))

sum = 0
i = 0
while (i <= n):
    sum = sum + i
    i = (i + 1)

print(sum)
```

```python
#Q4: Calculate factorial of a number using a loop.

n = int(input("Enter number :"))

factorial = 1

for i in range (1, n + 1):
    factorial = factorial * i

print("factorial of ",n,"is",factorial)
```

```python
#Q5: Print a pyramid pattern: *, **, ***, ****, *****

n = int(input("Enter a Number :"))

for i in range(1,n+1):
    print("*"*i)
```

```python
#Q7: Generate first 10 Fibonacci numbers.

#A Fibonacci series is a sequence of numbers where each number is the sum of the previous two numbers.

a = 0 
b = 1

for i in range (10):
    print(a)

    c = a + b
    a = b
    b = c
    



```

```python
#Q8: Print numbers 1-20, skip multiples of 3 (continue), stop at 15 (break).

for i in range (1,21):

    if i == 15:
        break 
    
    if i % 3 == 0 :
        continue
     
    print(i)



```

```python

#Q15: Calculate power of a number using loop.

base = int(input("Enter Base Number Here :"))
exponent = int(input("Enter Exponent Here :"))

power = 1

for i in range (exponent):
    power = base * power

print("Result :" ,power)
```

```python
#Q17: Count and print vowels in a string using loop.

string = str(input("Enter a String :"))

count = 0

for i in string:
    if i in "aeiou":
        print(i)
        count = count + 1
print("Count of Vowels is :",count)
```
